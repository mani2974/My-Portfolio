# Detecting the outliers

```python
#checking the outliers using boxplot:

##function to calculate q1, q2 , iqr and outliers in each column
def find_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_value = q1 - 1.5 * iqr
    upper_value =q3 + 1.5* iqr
    outliers = df[(df[column] < lower_value) | (df[column] > upper_value)]
    print(f'Q1 :{q1}')
    print(f'Q3 :{q3}')
    print(f'IQR:{iqr}')
    print(f'No of outliers:{outliers[column].shape[0]}')
    print(f'Unique outlier values of {column} column: {outliers[column].unique().tolist( )}')


##boxplot of each integer column to visualise outliers:
int_col = df.select_dtypes(include='number').columns
for i in int_col:
  find_outliers(df,i)
  plt.figure(figsize=(6,5))
  sns.boxplot(y = df[i], color = 'orange')
  plt.tight_layout()
  plt.show()
  print('\n')
```

# Probability calculations

### Probability calculation with gender and marital status

```python
#probability analysis of each product based on Gender:
df_cross = df.copy()
cross_g = pd.crosstab( index =[df_cross['Product']],columns=[df_cross['Gender'],df['MaritalStatus']]).reset_index()
cross_g.set_index('Product', inplace =True)
print(cross_g,'\n')

#marginal probability:
Probability_of_products = cross_g.iloc[:, :].sum(axis=1).div(cross_g.iloc[:,:].values.sum())
print(f'Probability of the users buying all products:â£\n{Probability_of_products.round(2)*100} \n')
Probability_of_gender = cross_g.iloc[:, 0:].sum(axis=0).div(cross_g.iloc[:,0:].values.sum())
print(f'Probability of male and female users buying all products :\n{Probability_of_gender.round(2)*100}\n')
```

**Insights:**

* KP281 (Entry-level) is the most purchased treadmill (44%).

* KP481 (Mid-range) follows at 33%.

* KP781 (High-end) has the least purchases at 22%.

* Female Customers:

    1. It is clear that partnered females are more likely to purchase KP281 (59%) and less likely to buy KP781 (9%).

    2. And also Single females prefer KP481 (47%) and KP281 (43%) over KP781 (10%).

* Male Customers:
    1. The partnered males are almost evenly split between KP281 (34%) and KP481 (34%), but a notable 31% go for KP781.

    2. Single males have a higher preference for KP781 (33%) and KP281 (44%).




**Suggestions:**
* KP281 (Entry Model):

   1. It is suggested for aerofit to focus marketing on partnered and single females, as they show strong preference for this model.

  2. it is recommended to target budget-conscious users or first-time treadmill buyers.

* KP481 (Mid-Range Model):

  1. It is a balanced choice for both males and females.

  2. It is a  value for money model and moderate usage features.

  3. It is seen that this model have strong interest among partnered males and single females.

* KP781 (High-End Model):

  1. It is usggested to focus this model primarily to males, especially partnered males and higher-income single males.

  2. It is recommended to highlight advanced features, durability, and performance for serious runners.

  3. This modelhave very low uptake among femalesâ€”consider repositioning or adjusting features if targeting females.




```python
#Conditional probability:
#P(Gender|Product)
Gender_given_product = cross_g[['Female', 'Male']].div(cross_g[['Female', 'Male']].sum(axis=1), axis=0)
print(f'Probability of each gender if product purchased is given:\n{Gender_given_product.round(2)*100}\n')
#P(product|Gender)
Product_given_Gender = cross_g.iloc[:, 0:].div(cross_g.iloc[:, 0:].sum())
print(f'Probability of each product purchased if gender is given:\n{Product_given_Gender.round(2)*100}\n')
```

**conditional probability insights:**

*  If KP781 is purchased:

    There's a high probability that the buyer is male (Partnered 48%, Single 35%).

    Very low chance it's a female (Partnered 10%, Single 8%).

* If KP281 is purchased:

    More likely to be a female (50% combined) than a male (50%), but with slightly higher count from partnered females.

* If KP481 is purchased:

    Partnered males (35%) and single females (23%) show strong preference.
